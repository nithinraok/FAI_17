```python
import numpy as np
import pandas as pd
import tensorflow as tf
# from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('sonar_data.csv',header=None)
```

```python
data.replace(to_replace='R',value=0,inplace=True)
data.replace(to_replace='M',value=1,inplace=True)
```

```python
data_x=data.get_values()[:,:-1];
data_y=data.get_values()[:,-1];
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x,data_y,test_size=0.25,random_state=10)
```

```python
train_y=train_y.reshape(-1,1);
test_y=test_y.reshape(-1,1)
```

```python
lr=0.01;
optimizer=tf.train.GradientDescentOptimizer(lr);
x=tf.placeholder(tf.float32,shape=[None,60]);
y=tf.placeholder(tf.float32,shape=[None,1]);

W1=tf.Variable(tf.zeros([60,10]),tf.float32);
b1=tf.Variable(tf.zeros([1,10]),tf.float32);
W2=tf.Variable(tf.zeros([10,1]),tf.float32);
b2=tf.Variable(tf.zeros([1,1]),tf.float32);

output1=tf.matmul(x,W1)+b1;
output2=tf.matmul(tf.nn.relu(output1),W2)+b2;
pred=tf.nn.sigmoid(output2);
init=tf.global_variables_initializer();

cost_func = tf.nn.sigmoid_cross_entropy_with_logits(logits=pred,labels=y);

model=optimizer.minimize(cost_func);
correct_prediction = tf.equal(tf.cast(pred + 0.5, tf.int32), tf.cast(y + 0.5, tf.int32));
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32));
```

```python
cost_history=[];
loss_history=[];
```

```python
sess=tf.Session();
sess.run(init);

for i in range(10+1):
    sess.run(model, feed_dict={x: train_x, y: train_y});
#     if i%100 == 0:
    print('Training Step:' + str(i) + '  Accuracy =  ' + str(sess.run(accuracy, feed_dict={x: test_x, y: test_y})) + '  Loss = ' + str(sess.run(cost_func, {x: train_x, y: train_y})))
#     graph = tf.summary.FileWriter("./graphs",sess.graph)
```
